```python
import numpy as np
import matplotlib.pyplot as plt
from ENOsmall import ENO
from CAPMr1 import CAPM
```

```python
TEST_LOCATION = 'tokyo'
TEST_YEAR = '2015'
```

```python
capm = CAPM(TEST_LOCATION,TEST_YEAR, shuffle=False, trainmode=False) #if reset is True
capm.eno = ENO(TEST_LOCATION,TEST_YEAR, shuffle=False, day_balance=False) #instantiate the environment inside the CAPM class
capm.HMAX = capm.eno.SMAX #maximum power output of solar cell is set in CAPM object using the value in ENO object
capm.reset();
```

```python
capm.batt = 0.5*capm.BMAX
```

```python
print('Battery = ',capm.batt/capm.BMAX)
print('Violation Flag = ',capm.violation_flag)
print('BMEAN = ', np.mean(capm.btrack))
```

```python
z0 = []
x = np.arange(0,capm.BMAX/2)
for capm.enp in x:
    z0 = np.append(z0,capm.rewardfn())

plt.plot(x,z0)
```

```python
z1 = []
y = np.arange(0,capm.BMAX/2)
for bmean in y:
    bdev = np.abs(capm.BOPT - bmean)/capm.BMAX
    if (bdev <= 0.1):
        penalty = 0
    else:
        VTh = 0.2
        penalty = np.exp(bdev/VTh)/np.exp(0.35/VTh) # max penalty is 1 when mean bdev = 0.35
    z1 = np.append(z1,penalty)
plt.plot(y,z1)
```

```python
def r1(enp):
    capm.enp = enp
    return capm.rewardfn()
vr1 = np.vectorize(r1)
```

```python
def r2(bmean):
    bdev = np.abs(capm.BOPT - bmean)/capm.BMAX
    if (bdev <= 0.2):
        penalty = -1
    else:
        VTh = 0.2
        penalty = np.exp(bdev/VTh)/np.exp(0.35/VTh) # max penalty is 1 when mean bdev = 0.35
    return penalty
vr2 = np.vectorize(r2)
```

```python
rangelist = np.arange(0,capm.BMAX/2,100)
listsize = len(rangelist)
z = np.zeros((listsize,listsize))
for enp in np.arange(listsize):
    for bmean in np.arange(listsize):
        z[enp][bmean] = r1(rangelist[enp])-r2(rangelist[bmean])
```

```python
plt.imshow(z, cmap='jet')
plt.colorbar()
plt.xlabel('ENP')
plt.ylabel('BMEAN')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import axes3d
import numpy

x = [i for i in range(0,4000,10)]
y = x

X, Y = numpy.meshgrid(x, y)
Z = vr1(X)-vr2(Y)

fig = plt.figure(figsize=(30,15))
ax = fig.add_subplot(111, projection = '3d')
surf = ax.plot_surface(X, Y, Z, rstride = 5, cstride = 5, cmap = matplotlib.cm.jet)
plt.xlabel('ENP')
plt.ylabel('BMEAN')
fig.colorbar(surf)
plt.show()
```

```python
def r(x):
    # based on the sigmoid function
    return (1/(1+np.exp(-x)))

x = np.arange(-5,5,0.01)
y = [r(i) for i in x]
plt.plot(x,y)
plt.grid(b=True, which='major', axis='both')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
```

```python
def r1(x):
    # based on the sigmoid function
    # x ranges from bdev = (0,0.5) of BMAX
    p_sharpness = 10
    n_sharpness = 20
    shift = 0.5
    # r1(x) = 0.5 when x = 0.25. Therefore, shift = 0.5 to make sure that (2*x-shift) evaluates to zero at x = 0.25
    if(x<=0.25): return 2*(1-(1 / (1 + np.exp(-p_sharpness*(2*x-shift)))))-1
    else: return 2*(1-(1 / (1 + np.exp(-n_sharpness*(2*x-shift)))))-1
```

```python
x = np.arange(0,0.5,0.01)
y = [r1(i) for i in x]
plt.plot(x,y)
plt.grid(b=True, which='major', axis='both')
plt.xlabel('|bdev|')
plt.ylabel('r1(|bdev|)')
```

```python
def r2(x):
    # based on the sigmoid function
    # x ranges from enp = (0,3) of DMAX
    p_sharpness = 2
    n_sharpness = 2
    shift = 6
    # r1(x) = 0.5 when x = 2. Therefore, shift = 6 to make sure that (3*x-shift) evaluates to zero at x = 2
    if(x<=2): return (1 / (1 + np.exp(p_sharpness*(3*x-shift))))
    else: return (1 / (1 + np.exp(n_sharpness*(3*x-shift))))
```

```python
x = np.arange(0,3,0.01)
y = [r2(i) for i in x]
plt.plot(x,y)
plt.grid(b=True, which='major', axis='both')
plt.xlabel('|enp|')
plt.ylabel('r2(|enp|)')
```
